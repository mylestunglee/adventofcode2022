```python
directions = ">>><<><>><<<>><>>><<<>>><<<><<<>><>><<>>"

with open('day17_input.txt') as file:
    directions = file.read()

shape1 = [
['#', '#', '#', '#']]

shape2 = [
['.', '#', '.'],
['#', '#', '#'],
['.', '#', '.']]

shape3 = [
['#', '#', '#'],
['.', '.', '#'],
['.', '.', '#']]

shape4 = [
['#'],
['#'],
['#'],
['#']]

shape5 = [
['#', '#'],
['#', '#']]
    
shapes = [shape1, shape2, shape3, shape4, shape5]

grid = [['.'] * 7 for _ in range(99999)]

def is_valid(shape, x, y):
    # check out of bounds
    shape_width = len(shape[0])
    shape_height = len(shape)
    grid_width = len(grid[0])
    if x < 0 or x + shape_width > grid_width or y < 0:
        return False
    
    # check overlap
    for j in range(shape_height):
        for i in range(shape_width):
            if grid[y + j][x + i] == '#' and shape[j][i] == '#':
                return False
            
    return True

def grid_height():
    y = 0
    while '#' in grid[y]:
        y += 1
    return y

def calc_column_height(x):
    y = 0
    while grid[y][x] == '#':
        y += 1
    return y

def calc_spawn_position(shape):
    return (2, grid_height() + 3)

def place_shape(shape, x, y):
    shape_width = len(shape[0])
    shape_height = len(shape)

    for j in range(shape_height):
        for i in range(shape_width):
            if shape[j][i] == '#':
                grid[y + j][x + i] = '#'

def print_grid():
    height = grid_height()
    for row in grid[0:height][::-1]:
        print('|' + ''.join(row) + '|')

shape_i = 0
direction_i = 0

for _ in range(2022):
    shape = shapes[shape_i]
    x, y = calc_spawn_position(shape)

    while True:
        if directions[direction_i] == '<' and is_valid(shape, x - 1, y):
            x -= 1
        elif directions[direction_i] == '>' and is_valid(shape, x + 1, y):
            x += 1
            
        direction_i = (direction_i + 1) % len(directions)
                
        if is_valid(shape, x, y - 1):
            y -= 1
        else:
            break


    place_shape(shape, x, y)

    shape_i = (shape_i + 1) % len(shapes)

print(grid_height())
```
